```python
# Import necessary libs
import streamlit as st
import pandas as pd
```

```python
st.title("Ethiopia Climate Finance Dashboard")

df = pd.read_parquet("../data/processed/ethiopian_climate_analysis.parquet")

st.subheader("Climate Finance Trend")

trend = df.groupby("Year")["USD_Commitment"].sum()
st.line_chart(trend)
```
